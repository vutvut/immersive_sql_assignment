```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
URL = "https://en.wikipedia.org/wiki/List_of_counties_in_Maryland"
```

```python
response = requests.get(URL)
response.status_code
```

```python
raw_HTML = BeautifulSoup(response.text, 'html.parser')
print(raw_HTML)
```

```python
raw_HTML.find('table', class_='wikitable sortable')
```

```python
county = {'FIPS Code': [],
          'County Seat': [], 
          'Established (year)': [],
          'Origin': [], 
          'Etymology': [], 
          'Flag': [],
          'Seal': [],
          'Population': [], 
          'Area': []}
```

```python
for row in tabledata:
    row_data = row.find_all('td')
    
    for index, item in enumerate(county.values()):
        data = row_data[index].text
  
        if (data == "\n") or (data == ''):
            continue
        else:
            item.append(data)
```
